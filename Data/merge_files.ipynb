```python
import pandas as pd
```

```python
f5 = pd.read_csv("2015.csv")
f5 = f5.drop(columns=['Region', 'Standard Error', 'Dystopia Residual'])
f5 = f5.rename(columns={"Economy (GDP per Capita)": "GDP per Capita", "Family": "Social support", "Health (Life Expectancy)": "Life Expectancy", "Freedom": "Freedom to make life choices", "Trust (Government Corruption)": "Government Trust"})
year = 2015
f5['Year'] = year
f5.head()
```

```python
f6 = pd.read_csv("2016.csv")
f6 = f6.drop(columns=['Region', 'Lower Confidence Interval', 'Upper Confidence Interval', 'Dystopia Residual'])
f6 = f6.rename(columns={"Economy (GDP per Capita)": "GDP per Capita", "Family": "Social support", "Health (Life Expectancy)": "Life Expectancy", "Freedom": "Freedom to make life choices", "Trust (Government Corruption)": "Government Trust"})
year = 2016
f6['Year'] = year
f6.head()
```

```python
f7 = pd.read_csv("2017.csv")
f7 = f7.drop(columns=['Whisker.high', 'Whisker.low', 'Dystopia.Residual'])
f7 = f7.rename(columns={"Happiness.Rank": "Happiness Rank","Happiness.Score": "Happiness Score", "Economy..GDP.per.Capita.": "GDP per Capita", "Family": "Social support", "Health..Life.Expectancy.": "Life Expectancy", "Freedom": "Freedom to make life choices", "Trust..Government.Corruption.": "Government Trust"})
year = 2017
f7['Year'] = year
columns_titles = ["Country", "Happiness Rank", "Happiness Score", "GDP per Capita", "Social support", "Life Expectancy", "Freedom to make life choices", "Government Trust", "Generosity", "Year"]
f7=f7.reindex(columns=columns_titles)
f7.head()
```

```python
f8 = pd.read_csv("2018.csv")
f8 = f8.rename(columns={"Overall rank": "Happiness Rank", "GDP per capita": "GDP per Capita", "Country or region": "Country", "Score": "Happiness Score", "Healthy life expectancy": "Life Expectancy", "Perceptions of corruption": "Government Trust"})
year = 2018
f8['Year'] = year

f8 = f8[["Country", "Happiness Rank", "Happiness Score", "GDP per Capita", "Social support", "Life Expectancy", "Freedom to make life choices", "Government Trust", "Generosity", "Year"]]
f8.head()
```

```python
f9 = pd.read_csv("2019.csv")
f9 = f9.rename(columns={"Overall rank": "Happiness Rank", "GDP per capita": "GDP per Capita", "Country or region": "Country", "Score": "Happiness Score", "Healthy life expectancy": "Life Expectancy", "Perceptions of corruption": "Government Trust"})
year = 2019
f9['Year'] = year
columns_titles = ["Country", "Happiness Rank", "Happiness Score", "GDP per Capita", "Social support", "Life Expectancy", "Freedom to make life choices", "Government Trust", "Generosity", "Year"]
f9=f9.reindex(columns=columns_titles)
f9.head()
```

```python
frames = [f5, f6, f7, f8, f9]
result = pd.concat(frames)
result.head()
```
